```python
11. In your own words, list some advantages and disadvantages of BMI, waist-to-height 
ratio, and waist-to-hip ratio.
```

```python
Advantage: 
1.Simple and Quick:
Only requires height and weight, making it easy to compute.
2.Widely Recognized:
Commonly used globally to classify individuals into underweight, normal, overweight, and obese categories.
3.Effective for Population Studies:
Useful for identifying trends related to obesity in large groups of people.
4.Predictive Value:
Correlates with risk factors for diseases like diabetes, cardiovascular disease, and hypertension.
      
Disadvantages:
1.Ignores Body Composition:
Cannot distinguish between fat mass and lean muscle mass. A muscular person may have a high BMI but low body fat.
2.Ignores Fat Distribution:
BMI does not account for where fat is located in the body (e.g., visceral vs. subcutaneous fat).
3.Not Universal:
May not work well for children, older adults, or different ethnic groups.
4.False Positives/Negatives:
Someone can be classified as "normal" BMI but still have high body fat or risk factors for disease.
```
